1. 질문 :  검색어로  찾은  책 목록을 json 파일로 저장하기
   data/books.json 파일로 저장해 주세요.

```python
import requests
from pprint import pprint
import os
import json
from dotenv import load_dotenv

# .env 파일에서 환경 변수 로드
load_dotenv()

# 환경 변수에서 NAVER API 인증 정보 불러오기
client_id = os.getenv("NAVER_CLIENT_ID")
client_secret = os.getenv("NAVER_CLIENT_SECRET")

# 공통 headers 설정
headers = {
    'X-Naver-Client-Id': client_id,
    'X-Naver-Client-Secret': client_secret,
}

#책 검색 함수
def search_books(query):
    url = 'https://openapi.naver.com/v1/search/book.json'
    payload = {
        'query': query,
        'display': 50,  # 최대 100까지 가능
        'sort': 'sim'
    }

    res = requests.get(url, params=payload, headers=headers)

    if res.status_code == 200:
        items_data = res.json()['items']
        print(f"'{query}' 검색 결과: {len(items_data)}건")
        return items_data
    else:
        print(f"오류 발생: {res.status_code}")
        return []

# 결과를 JSON 파일로 저장하는 함수
def save_to_json(data, filename='data/books.json'):
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
    print(f"검색 결과가 '{filename}'에 저장되었습니다.")

# 실행 부분
if __name__ == "__main__":
    query = input("검색할 책 제목을 입력하세요: ")
    result = search_books(query)
    if result:
        save_to_json(result)
        pprint(result[0])  
```

4. 질문 :  검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기

title,author,discount,publisher,pubdate 컬럼만 출력
가격은 descending (내림차순), index 초기화


```python
book_df.loc[book_df['discount'] >= 20000, ['title', 'author', 'discount', 'publisher', 'pubdate']]\
.sort_values(by='discount',ascending=False)\
.reset_index(drop=True)
```
